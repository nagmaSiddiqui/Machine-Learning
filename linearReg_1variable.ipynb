######Linear Regression with single variable######

```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv('/content/insurance_dataset - Sheet1.csv')
```

```python
df
```

```python
import seaborn as sns
sns.lmplot(x='Age',y='Premium',data=df)
```

```python
reg=linear_model.LinearRegression()
```

```python
#check column name and clean them
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
reg.fit(df[['Age']], df['Premium'])
```

```python
reg.predict([[21]])
```

```python
reg.predict([[50]])
```

```python
reg.coef_
```
